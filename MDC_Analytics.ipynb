```python
"""Check out the Trending youtube video statistics.Analyze trends in the YouTube
video statistics to gain insights into what makes a video trend."""
import pandas as pd
try:
    df_in = pd.read_csv('INvideos.csv', encoding='utf-8')
    df_ca = pd.read_csv('CAvideos.csv', encoding='utf-8')
    df_kr = pd.read_csv('KRvideos.csv', encoding='ISO-8859-1')  # got error with previous encoding so, trying a different encoding
    df_gb = pd.read_csv('GBvideos.csv', encoding='utf-8')
    df_us = pd.read_csv('USvideos.csv', encoding='utf-8')
except UnicodeDecodeError as e:
    print(f"UnicodeDecodeError: {e}")
#adding Country/state column in evry data set to compare various attributes w.r.t country.
df_in['Country']='India'
df_ca['Country']='Calfornia'
df_kr['Country']='South Korea'
df_gb['Country']='Britain'
df_us['Country']='US'
# Combine dataframes
df_list = [df_ca, df_in, df_kr, df_gb, df_us]
df_combined = pd.concat(df_list)
```

# **To plot how is each country having more views in some particular category_ids than others:**

```python
#a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=df_aggregated2, x='Country', y='views', hue='category_id', palette='Set1')
# Add plot title and labels
plt.title('Views by Country and Category ID', fontsize=16)
plt.xlabel('Category ID', fontsize=14)
plt.ylabel('Views', fontsize=14)
plt.legend(title='category_id')
plt.xticks(rotation=45)  # Rotating x-axis labels
plt.tight_layout()
plt.show()
```

### **To plot how are no.of likes varying w.r.t category_ids:**

```python
# a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_combined, x='category_id', y='likes', hue='category_id', palette='Set3', s=100, edgecolor='w')
# Add plot title and labels
plt.title('Scatter Plot of Number of Likes by Category ID', fontsize=16)
plt.xlabel('Category ID', fontsize=14)
plt.ylabel('Number of Likes', fontsize=14)
plt.xticks(ticks=df_combined['category_id'].unique(), rotation=45)  # Rotate x-axis labels
#unique()- above, using to have each and evry category_id on x axis, so that it's respective scattering in no.of likes is showed.
plt.legend(title='Category ID',bbox_to_anchor=(1.05, 1), loc='upper left')
# bbox_to_anchor=(1.05, 1)- to put the category_id box(legend) to the right most side
#1.05 indicates that the legend box should be placed slightly to the right of the plot (5% beyond the right edge of the plot).
#1 indicates that the legend box should be aligned with the top edge of the plot.
plt.tight_layout()
plt.show()
```

# **Analyzing the distribution of errors or removed videos across each country**

```python
country_data = df_combined[df_combined['Country'] == 'US']
# to count occurrences of True and False
error_counts = country_data['video_error_or_removed'].value_counts()
# Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(error_counts, labels=error_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
# Add plot title
plt.title('Distribution of Video Error or Removed for USA', fontsize=16)
# Show the plot
plt.show()
```
